#### Example running SpaFHy with jupyter notebook

```python
from model_driver import driver
from iotools import read_results
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
import os

os.chdir(r'C:\SpaFHy_v1_Pallas_2D')

# example of calling driver, reading results
outputfile = driver(create_ncf=True, create_spinup=False, folder='testcase_input')
results = read_results(outputfile)
```

```python
results['bucket_moisture_top'].sel(time=slice('2021-06-01', '2021-09-01')).mean(dim=['lat','lon']).plot(label='organic moss-humus layer')
results['bucket_moisture_root'].sel(time=slice('2021-06-01', '2021-09-01')).mean(dim=['lat','lon']).plot(label='rootzone layer')
plt.legend()
```

```python
results['bucket_moisture_root'].sel(time=slice('2021-05-01', '2021-09-01')).mean(dim=['time']).plot(cmap='coolwarm_r', vmin=0.0, vmax=1.0, label='rootzone layer')
```
